### Pretty crude methodology of obtaining representative zip codes for top metropolitan areas in the US
### I refer to a wikipedia article listing top metropolitan areas, then search for the most populous zip code for that region in a dataframe which contains all US zip codes


```python
import numpy as np
import pandas as pd
```

```python
DF = pd.read_csv('free-zipcode-database-Primary.csv')
```

```python
DF.head()
```

```python
DF.fillna(0).sort_values(by='EstimatedPopulation')
```

```python
DF.EstimatedPopulation.mean()
```

```python
DF.fillna(DF.EstimatedPopulation.mean())[DF.City=='NEW YORK'].sort_values(by='EstimatedPopulation')
```

```python
import matplotlib.pyplot as plt
```

```python
DF.fillna(DF.EstimatedPopulation.mean())[DF.City=='NEW YORK'].sort_values(by='EstimatedPopulation').tail(1).Zipcode.values[0]
```

```python
def get_zip_of_city(thecity):
    print(DF.fillna(DF.EstimatedPopulation.mean())[DF.City==thecity].sort_values(by='EstimatedPopulation').tail(1).Zipcode.values[0])
```
